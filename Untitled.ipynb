```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
train_feature = pd.read_csv(r"C:\Users\malvi\training_set_features.csv")
train_target=pd.read_csv(r"C:\Users\malvi\training_set_labels.csv")
df_test = pd.read_csv(r"C:\Users\malvi\test_set_features.csv")
submission=pd.read_csv(r"C:\Users\malvi\submission_format.csv")
```

```python
df = train_feature.merge(train_target,on="respondent_id")
```

```python
df.set_index("respondent_id",inplace=True)
df.head()
```

```python
categories=df.columns
categories
```

```python
for x in categories:
    cat_mode = df[x].mode()[0]
    df[x].fillna(cat_mode, inplace=True)
    
```

```python
df.drop_duplicates(keep=False,inplace=True)
```

```python
df.duplicated().any()
```

```python
categorical_df=df.select_dtypes(include="object")
categorical=categorical_df.columns
categorical
```

```python
for x in categorical:
    one_hot_encoder = OneHotEncoder()
    one_hot_encoded = one_hot_encoder.fit_transform(df[[x]]).toarray()

    # Convert the result to a DataFrame and concatenate with the original DataFrame
    one_hot_encoded_df = pd.DataFrame(one_hot_encoded, columns=one_hot_encoder.get_feature_names_out([x]))
    one_hot_encoded_df.index = df.index
    df = pd.concat([df, one_hot_encoded_df], axis=1)
    df.drop(x, axis=1, inplace=True)
```

```python
df_test.set_index("respondent_id",inplace=True)
df_test.head()
```

```python
categories=df_test.columns
categories
```

```python
for x in categories:
    cat_mode = df_test[x].mode()[0]
    df_test[x].fillna(cat_mode, inplace=True)
```

```python
df_test.isnull().sum()
```

```python
categorical_df = df_test.select_dtypes(include="object")
categorical = categorical_df.columns
categorical
```

```python
for x in categorical:
    one_hot_encoder = OneHotEncoder()
    one_hot_encoded = one_hot_encoder.fit_transform(df_test[[x]]).toarray()

    # Convert the result to a DataFrame and concatenate with the original DataFrame
    one_hot_encoded_df = pd.DataFrame(one_hot_encoded, columns=one_hot_encoder.get_feature_names_out([x]))
    one_hot_encoded_df.index = df_test.index
    df_test = pd.concat([df_test, one_hot_encoded_df], axis=1)
    df_test.drop(x, axis=1, inplace=True)
```

```python
X_train = df.drop(["xyz_vaccine","seasonal_vaccine"],axis=1)
Y_train_xyz = df[["xyz_vaccine"]]
Y_train_seas = df[["seasonal_vaccine"]]
```

```python
model_xyz=LogisticRegression(random_state=42,max_iter=10000)
model_xyz.fit(X_train,Y_train_xyz)
```

```python
model_seas=LogisticRegression(random_state=42,max_iter=10000)
model_seas.fit(X_train,Y_train_seas)
```

```python
predict_xyz=model_xyz.predict_proba(df_test)[:,1]
predict_xyz
```

```python
predict_seas=model_seas.predict_proba(df_test)[:,1]
predict_seas
```

```python
submissions = pd.DataFrame({
    'respondent_id': submission['respondent_id'],
    'xyz_vaccine': predict_xyz,
    'seasonal_vaccine': predict_seas
})
```
